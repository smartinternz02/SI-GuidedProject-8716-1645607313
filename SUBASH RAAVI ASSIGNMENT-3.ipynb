```python
# linegraph
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
x=[1,2,3,4,5]
y=np.power(x,2)
plt.plot(x,y)
```

```python
#Scatter plot
plt.scatter(x,y,color='green',marker='^')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Bar plot
plt.bar(x,y,width=.5,color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title("Bar Plot")
```

```python
# Histogram
plt.hist(x)
```

```python
#Count plot
import seaborn as sns
x=[120,80]
sns.countplot(x)
```

```python
# Heatmap
fig=pd.DataFrame([[1,20,30],[20,2,23],[23,34,1]])
sns.heatmap(fig,linewidth=1)
```

```python
# Boxplot
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.plot.box()
```

```python
# Pairplot
sns.pairplot(fig)
```

```python
# Jointplot  
data = sns.load_dataset("attention")
sns.jointplot(x = "solutions", y = "score",kind = "hex", data = data)
plt.show()
```

```python
# Kernel density plot
tips=sns.load_dataset("tips")
sns.kdeplot(data=tips,x="total_bill")
```
